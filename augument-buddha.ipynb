```python
images = [root_directory + img for img in os.listdir(root_directory)]
train_data, test_data = train_test_split(images, test_size = 0.2)
test_data
```

```python
all_transforms = [
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.15, saturation=0.2, hue=0.1),
    transforms.RandomGrayscale()
    ]
```

```python
def apply_transform(transform, data, isTrain, i, j):
    for image in data:
        img = Image.open(image).convert("RGB")
        img = transforms.ToTensor()(img)
        transformed_image = transform(img)
        transformed_image = transformed_image.clamp(0, 1)
#         download_img(transformed_image, isTrain
        transformed_image_pil = transforms.ToPILImage()(transformed_image)
        if(isTrain):
            transformed_image_pil.save(train_img_dir + f"transformed_{i}.jpg")
            i+=1
        else:
            transformed_image_pil.save(test_img_dir + f"transformed_{j}.jpg")
            j+=1
```

```python
mean = 0.0
std = 0.1
```

```python
def insert_noise(data, isTrain, i, j):
    for image in data:
        img = Image.open(image).convert("RGB")
        img = transforms.ToTensor()(img)
        noise = torch.randn(img.size()) * std + mean
        noisy_img = img + noise
        noisy_img = torch.clamp(noisy_img, 0, 1)
        transformed_image_pil = transforms.ToPILImage()(noisy_img)
        if(isTrain):
            transformed_image_pil.save(train_img_dir + f"transformed_{i}.jpg")
            i+=1
        else:
            transformed_image_pil.save(test_img_dir + f"transformed_{j}.jpg")
            j+=1
```

```python
i = 1
j = 1
for transform in all_transforms:
    apply_transform(transform, train_data, True, i, j)
    i+=len(train_data)
    apply_transform(transform, test_data, False, i, j)
    j+=len(test_data)
insert_noise(train_data, True, i, j)
insert_noise(test_data, True, i, j)
```
